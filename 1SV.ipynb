```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.model_selection import train_test_split

from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.optimizers import Adam
from keras.losses import CategoricalCrossentropy, MeanAbsoluteError
from sklearn.preprocessing import StandardScaler
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
N = 30
Img_data = np.zeros((N*3,150,150), dtype=np.uint8)
#Chuyển ảnh thành mảng
for i in range(0,N): 
  img = cv2.imread('NHT'+str(i)+'.jpg',cv2.IMREAD_GRAYSCALE)
  Img_data[i][::][::] = img
for i in range(0,N): 
  img = cv2.imread('DTDH'+str(i)+'.jpg',cv2.IMREAD_GRAYSCALE)
  Img_data[N+i][::][::] = img
for i in range(0,N): 
  img = cv2.imread('DCT'+str(i)+'.jpg',cv2.IMREAD_GRAYSCALE)
  Img_data[2*N+i][::][::] = img
```

```python
#Tạo mảng chưa tên, 0 là NHT, 1 là DTDH, 2 là DCT
Name = np.zeros((N*3), dtype=np.uint8)
for i in range(0,30):
  Name[i] = 0
for i in range(30,90):
  Name[i] = 1
```

```python
labels = [['Đay là Nguyễn Hồng Thuận'],['Không phải Nguyễn Hồng Thuận']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(Img_data,Name,test_size = 0.3)
x_train = x_train.reshape(63,150*150)
x_test = x_test.reshape(27,150*150)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train/=255.0
x_test/=255.0
y_train = np_utils.to_categorical(y_train,2)
y_test = np_utils.to_categorical(y_test,2)
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(22500,)))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(2, activation= 'softmax'))
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
history = model.fit(x_train,y_train,batch_size = 2, epochs = 50,verbose = 1, validation_data =(x_test,y_test))
```

```python
model.save('Nhandien1SVc.h5')
```

```python
name_img = 'NHT_Test.jpg'
Img_data = cv2.imread(name_img,cv2.IMREAD_GRAYSCALE)
Img_data = Img_data.reshape(1,150*150)
Img_data = Img_data.astype('float32')
Img_data/=255.0
Img_show = cv2.imread(name_img,cv2.IMREAD_COLOR)
Img_show = cv2.cvtColor(Img_show, cv2.COLOR_BGR2RGB)
plt.imshow(Img_show)
print(labels[np.argmax(model.predict(Img_data))])
```
